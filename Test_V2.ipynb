```python
from configs import PipelineConfig, SuperResolutionConfig
from pipeline import LPRPipeline
from PIL import Image

# Create config
config = PipelineConfig(
    debug=True,
    start_stage="car_detection",
    end_stage="ocr",
    super_resolution=SuperResolutionConfig(
        apply_threshold=True,
        min_width=60,
        min_height=20
    )
)

# Create pipeline
pipeline = LPRPipeline(config)

# Process image
image = Image.open("./DATA_Test/day_00141.jpg")
results = pipeline.process(image)

# Get results
for result in results:
    print(f"Detected {result.plates_detected} plates")
    for text in result.recognized_texts:
        print(f"  - {text}")
```

```python
import torch

# Path checkpoint
checkpoint_path = "weights/super_resolution.pth"

checkpoint = torch.load(checkpoint_path, map_location='cpu')

if isinstance(checkpoint, dict):
    print(f"\nCheckpoint keys: {list(checkpoint.keys())}")

    if 'model_state_dict' in checkpoint:
        state_dict = checkpoint['model_state_dict']
        print("\nUsing 'model_state_dict' key")
    elif 'state_dict' in checkpoint:
        state_dict = checkpoint['state_dict']
        print("\nUsing 'state_dict' key")
    else:
        state_dict = checkpoint
        print("\nCheckpoint appears to be a state_dict directly")
else:
    state_dict = checkpoint
    print("\nCheckpoint is a state_dict directly")

print(f"\nTotal keys in state_dict: {len(state_dict)}")
print("\nAll keys in state_dict:")
for i, key in enumerate(state_dict.keys()):
    print(f"  {i+1:3d}. {key}")


print(f"  'mean' in state_dict: {'mean' in state_dict}")
print(f"  'std' in state_dict: {'std' in state_dict}")

for i, (key, value) in enumerate(list(state_dict.items())[:10]):
    print(f"  {key}: {value.shape}")
```
